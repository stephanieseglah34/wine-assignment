```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
red_wine_csv = "red_wine.csv"
red_wine_df = pd.read_csv(red_wine_csv)
```

```python
# Preview first 10 rows
red_wine_df.head(10)
```

```python
# Preview last 20 rows
red_wine_df.tail(20)
```

```python
# Use info method
red_wine_df.info()
```

```python
# Statistical info about the data
red_wine_df.describe()
```

```python
red_wine_df.hist(bins=25, figsize=(10,10))
plt.show()
```

**Analysis of the above histogram plots**
* alcohol has a near good distribution.
* sulphates is skewed to the right.
* The rest do not have a normal distribution

```python
# Histogram of alcohol content
plt.figure(figsize=(8, 6))
plt.hist(red_wine_df['alcohol'], bins=30, edgecolor='black')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.title('Distribution of alcohol content')
plt.show()
```

From the chart above, we can see that the alcohol content of 9.6 has the heighest frequency.

```python
plt.figure(figsize=(20, 15))
sns.heatmap(red_wine_df.corr(), vmin=-1, vmax=1, cmap='coolwarm', annot=True)
plt.show()
```

**The following is noticed from the heatmap**
* A fair correlation exist between alcohol and quality
* High correlation exist betwen fixed acidity and citric acid and also density.
* There's a very low correlation between fixed acidity and pH
